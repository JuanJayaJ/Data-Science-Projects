```python
# Stationarity Test
results = {}
for crop in yield_columns:
    ts = data.groupby('Year').apply(lambda group: group[crop].multiply(group["Yield_hg/ha"]).sum())
    result = adfuller(ts)
    results[crop] = result[1] # p-value

# Print the p-values for the ADF test for each crop
for crop, p_value in results.items():
    print(f'{crop}: p-value = {p_value}')
```

#### The yield time series of these crops have very high P-values, much higher than 0.05, which strongly suggests that these time series are non-stationary.

# Linear Regression & Polynomial Regression

```python
# Importing required libraries
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Define features and target from the provided code
features = [
    'Item_Maize', 'Item_Plantains', 'Item_Potatoes', 'Item_Rice', 'Item_Sorghum', 'Item_Soya beans',
    'Item_Sweet potatoes', 'Item_Wheat', 'Item_Yams', 'CH4_kt', 'CO2_kt', 'CO2eq_kt', 'N2O_kt',
    'TotalGHG_MtCO2eq', 'AvgPrecipitation_mm/year', 'AvgTemp_DegC', 'PesticidesTotal_kg/ha', 
    'TempChange_DegC', 'Insecticides_t', 'NutrientNitrogenTotal_kg/ha'
]
target_column = 'Yield_hg/ha'

# Proceeding with the previous steps
data_subset = data[features + [target_column]]

# Split data into features and target
X = data_subset[features]
y = data_subset[target_column]

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Train the Linear regression model
linear_reg = LinearRegression()
linear_reg.fit(X_train, y_train)

# Train the Polynomial regression (degree 2) model
poly_reg = make_pipeline(PolynomialFeatures(degree=2), LinearRegression())
poly_reg.fit(X_train, y_train)

# Predictions for test set
linear_pred_test = linear_reg.predict(X_test)
poly_pred_test = poly_reg.predict(X_test)

# Predictions for training set
linear_pred_train = linear_reg.predict(X_train)
poly_pred_train = poly_reg.predict(X_train)

# Evaluation for test set
mse_linear_test = mean_squared_error(y_test, linear_pred_test)
mse_poly_test = mean_squared_error(y_test, poly_pred_test)

rmse_linear_test = mean_squared_error(y_test, linear_pred_test, squared=False)
rmse_poly_test = mean_squared_error(y_test, poly_pred_test, squared=False)

mae_linear_test = mean_absolute_error(y_test, linear_pred_test)
mae_poly_test = mean_absolute_error(y_test, poly_pred_test)

r2_linear_test = r2_score(y_test, linear_pred_test)
r2_poly_test = r2_score(y_test, poly_pred_test)

# Evaluation for training set
mse_linear_train = mean_squared_error(y_train, linear_pred_train)
mse_poly_train = mean_squared_error(y_train, poly_pred_train)

rmse_linear_train = mean_squared_error(y_train, linear_pred_train, squared=False)
rmse_poly_train = mean_squared_error(y_train, poly_pred_train, squared=False)

mae_linear_train = mean_absolute_error(y_train, linear_pred_train)
mae_poly_train = mean_absolute_error(y_train, poly_pred_train)

r2_linear_train = r2_score(y_train, linear_pred_train)
r2_poly_train = r2_score(y_train, poly_pred_train)

# Compile evaluation results into a DataFrame
eval_results = {
    "Metric": ["MSE", "RMSE", "MAE", "R^2"],
    "Linear Regression (Test)": [mse_linear_test, rmse_linear_test, mae_linear_test, r2_linear_test],
    "Polynomial Regression (Test)": [mse_poly_test, rmse_poly_test, mae_poly_test, r2_poly_test],
    "Linear Regression (Train)": [mse_linear_train, rmse_linear_train, mae_linear_train, r2_linear_train],
    "Polynomial Regression (Train)": [mse_poly_train, rmse_poly_train, mae_poly_train, r2_poly_train]
}

evaluation_df = pd.DataFrame(eval_results)
evaluation_df
```

```python
from sklearn.model_selection import KFold, cross_val_score

# Define a function to perform cross-validation for 1st and 2nd degree polynomial regression
def evaluate_poly_degree(degree, X, y, n_splits=5):
    model = make_pipeline(PolynomialFeatures(degree=degree), LinearRegression())
    scores = cross_val_score(model, X, y, cv=KFold(n_splits=n_splits, shuffle=True, random_state=42), scoring='neg_mean_squared_error')
    return -scores.mean()  # Return positive MSE

# Evaluate 1st and 2nd degree polynomial regression
degrees = [1, 2]
mse_scores = []

for degree in degrees:
    mse = evaluate_poly_degree(degree, X_train, y_train)
    mse_scores.append(mse)

# Find the best degree based on MSE
best_degree = degrees[mse_scores.index(min(mse_scores))]

# Train the best model on the full training data and evaluate on the test set
best_model = make_pipeline(PolynomialFeatures(degree=best_degree), LinearRegression())
best_model.fit(X_train, y_train)
best_predictions = best_model.predict(X_test)

# Evaluation metrics for the best model
mse_best = mean_squared_error(y_test, best_predictions)
rmse_best = mean_squared_error(y_test, best_predictions, squared=False)
mae_best = mean_absolute_error(y_test, best_predictions)
r2_best = r2_score(y_test, best_predictions)

eval_results_best = {
    "Metric": ["MSE", "RMSE", "MAE", "R^2"],
    f"Best Model (Degree {best_degree})": [mse_best, rmse_best, mae_best, r2_best]
}

evaluation_best_df = pd.DataFrame(eval_results_best)

degrees, mse_scores, best_degree, evaluation_best_df
```

```python
from sklearn.model_selection import KFold, cross_val_score
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Define a function for cross-validation
def evaluate_poly_degree(degree, X, y, n_splits=5):
    # Create a pipeline that first applies polynomial transformation and then linear regression
    model = make_pipeline(PolynomialFeatures(degree=degree), LinearRegression())
    # Use cross-validation to get the negative mean squared error for each split
    scores = cross_val_score(model, X, y, cv=KFold(n_splits=n_splits, shuffle=True, random_state=42), 
                             scoring='neg_mean_squared_error')
    return -scores.mean()  # Return the positive MSE

# Cross-validate the 2-degree polynomial model
degree = 2
mse = evaluate_poly_degree(degree, X_train, y_train)

# Train the 2-degree polynomial model on the entire training data and evaluate on the test set
model = make_pipeline(PolynomialFeatures(degree=degree), LinearRegression())
model.fit(X_train, y_train)
predictions = model.predict(X_test)

# Calculate MSE on the test set
mse_test = mean_squared_error(y_test, predictions)

mse, mse_test
```

This is because the polynomial model takes into account higher-degree terms and interactions among features, which can influence the weight of individual features. For instance, the coefficient for Item_Potatoes is notably high in the polynomial model, potentially due to the sparsity of potato data and its interaction effects with other features.

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
import numpy as np

# Prepare data for regression
years = yearly_production.index.values.reshape(-1, 1)
production_values = yearly_production.values

# Linear Regression
linear_regressor = LinearRegression()
linear_regressor.fit(years, production_values)
linear_predictions = linear_regressor.predict(np.arange(1990, 2031).reshape(-1, 1))

# Polynomial Regression (degree 2)
polynomial_features = PolynomialFeatures(degree=2)
years_poly = polynomial_features.fit_transform(years)
poly_regressor = LinearRegression()
poly_regressor.fit(years_poly, production_values)

# Predict using polynomial regressor
years_poly_pred = polynomial_features.transform(np.arange(1990, 2031).reshape(-1, 1))
poly_predictions = poly_regressor.predict(years_poly_pred)

# Plotting
plt.figure(figsize=(15, 7))
plt.plot(yearly_production.index, yearly_production.values, label='Actual Data', marker='o')
plt.plot(np.arange(1990, 2031), linear_predictions, label='Linear Regression', linestyle='--')
plt.plot(np.arange(1990, 2031), poly_predictions, label='Polynomial Regression (Degree 2)', linestyle='-.')
plt.title('Total Crop Production Over the Years with Predictions')
plt.ylabel('Total Production (adjusted by Yield)')
plt.xlabel('Year')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np

# Define the columns related to crop items
crop_columns = [col for col in data.columns if "Item_" in col]

# Adjust the total production for each crop item using the 'Yield_hg/ha' column
data['Adjusted_Production'] = 0
for crop in crop_columns:
    data['Adjusted_Production'] += data[crop] * data['Yield_hg/ha']

# Lists of developed and developing countries
developed_countries = ["United States", "Canada", "United Kingdom", "Germany", "France", 
                       "Japan", "Australia", "Sweden", "Switzerland", "Singapore"]
developing_countries = ["China", "India", "Brazil", "Mexico", "South Africa", "Nigeria", 
                        "Indonesia", "Turkey", "Vietnam", "Thailand"]

# Separate the data based on the adjusted production for developed and developing countries
developed_data_adjusted = data[data['Country Name'].isin(developed_countries)]
developing_data_adjusted = data[data['Country Name'].isin(developing_countries)]

# Function to perform regression and plotting
def perform_regression(data, title):
    # Prepare data for regression
    years = data['Year'].values.reshape(-1, 1)
    production_values = data['Adjusted_Production'].values
    
    # Linear Regression
    linear_regressor = LinearRegression()
    linear_regressor.fit(years, production_values)
    linear_predictions = linear_regressor.predict(np.arange(1990, 2031).reshape(-1, 1))
    
    # Polynomial Regression (degree 2)
    polynomial_features = PolynomialFeatures(degree=2)
    years_poly = polynomial_features.fit_transform(years)
    poly_regressor = LinearRegression()
    poly_regressor.fit(years_poly, production_values)
    
    # Predict using polynomial regressor
    years_poly_pred = polynomial_features.transform(np.arange(1990, 2031).reshape(-1, 1))
    poly_predictions = poly_regressor.predict(years_poly_pred)
    
    # Plotting
    plt.figure(figsize=(15, 7))
    plt.plot(data['Year'], data['Adjusted_Production'], label='Actual Data', marker='o')
    plt.plot(np.arange(1990, 2031), linear_predictions, label='Linear Regression', linestyle='--')
    plt.plot(np.arange(1990, 2031), poly_predictions, label='Polynomial Regression (Degree 2)', linestyle='-.')
    plt.title(title)
    plt.ylabel('Total Production (adjusted by Yield)')
    plt.xlabel('Year')
    plt.legend()
    plt.grid(True)
    plt.show()

# Perform regression for developed countries with adjusted production
perform_regression(developed_data_adjusted.groupby('Year').sum().reset_index(), 'Total Crop Production Over the Years (Developed Countries, adjusted by Yield)')

# Perform regression for developing countries with adjusted production
perform_regression(developing_data_adjusted.groupby('Year').sum().reset_index(), 'Total Crop Production Over the Years (Developing Countries, adjusted by Yield)')
```
